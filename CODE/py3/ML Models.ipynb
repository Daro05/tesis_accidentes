## Decision Tree Classifier

### Entrenamos el modelo DT par√°metros default con datos acumulados 2014-2018

```python
#Usando validaci√≥n cruzada con DT
instancia_kfold = KFold(n_splits = 10)
acc = cross_val_score(tree.DecisionTreeClassifier(random_state = 42), X,y, cv = instancia_kfold)
roc = cross_val_score(tree.DecisionTreeClassifier(random_state = 42), X,y, scoring = 'roc_auc',cv = instancia_kfold)
matt = cross_val_score(tree.DecisionTreeClassifier(random_state = 42), X,y, scoring = make_scorer(matthews_corrcoef),cv = instancia_kfold)
cohen = cross_val_score(tree.DecisionTreeClassifier(random_state = 42), X,y, scoring = make_scorer(cohen_kappa_score),cv = instancia_kfold)

print("acc",acc)
print("roc",roc)
print("matt",matt)
print("kappa",cohen)
```

### Corremos predicci√≥n para clasificaci√≥n de hex√°gonos para datos 2019

```python
clf = tree.DecisionTreeClassifier(random_state = 42)
   
clf = clf.fit(X, y)

y_pred = clf.predict(X_test19)
print(accuracy_score(y_test19, y_pred))
print("Kappa: ",cohen_kappa_score(y_test19, y_pred))
print("ROC_AUC_CURVE: ",roc_auc_score(y_test19, y_pred.round()))
print("MCC: ", matthews_corrcoef(y_test19, y_pred))
print(confusion_matrix(y_test19, y_pred))
```
